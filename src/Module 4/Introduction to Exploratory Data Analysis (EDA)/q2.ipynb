```python
# Write your code from here

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- Sample Datasets (including potential outliers) ---
employee_data = {'Employee ID': [101, 102, 103, 104, 105, 106],
                 'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
                 'Age': [25, 30, 28, 35, 22, 65],  # Potential outlier
                 'Department': ['Sales', 'Marketing', 'Engineering', 'Sales', 'HR', 'Management'],
                 'Salary': [50000, 60000, 75000, 55000, 48000, 90000],
                 'Joining Date': ['2023-01-15', '2022-05-20', '2024-03-01', '2023-08-10', '2024-01-25', '2021-06-01']}
df_employee = pd.DataFrame(employee_data)

sales_data = {'Product ID': [1, 2, 3, 4, 5, 6],
              'Product Name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam', 'Luxury Watch'],
              'Price': [1200.50, 25.00, 75.00, 300.00, 50.00, 5000.00],  # Potential outlier
              'Quantity Sold': [10, 50, 30, 15, 40, 5],
              'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Electronics', 'Luxury'],
              'Sales Date': ['2024-05-01', '2024-05-01', '2024-05-02', '2024-05-02', '2024-05-03', '2024-05-03']}
df_sales = pd.DataFrame(sales_data)

grades_data = {'Student ID': [1, 2, 3, 4, 5, 6],
               'Student Name': ['John', 'Jane', 'Peter', 'Mary', 'David', 'Sarah'],
               'Math Score': [85, 92, 78, 95, 88, 50],  # Potential outlier
               'Science Score': [78, 85, 90, 82, 76, 95],
               'English Score': [90, 88, 82, 94, 86, 70],
               'Year': [2023, 2023, 2024, 2024, 2023, 2024]}
df_grades = pd.DataFrame(grades_data)

# --- Task 1: Age in Employee Dataset ---
plt.figure(figsize=(6, 4))
sns.boxplot(x=df_employee['Age'])
plt.title('Box Plot of Age (Employee Dataset)')
plt.xlabel('Age')
plt.show()

print("\n--- Handling Outliers in Age (Employee Dataset) ---")
print("Potential Outlier Detected (visually above the upper whisker).")
print("Handling Methods:")
print("- **Removal:** If the outlier is a genuine error or represents an extremely rare case that is not representative of the population, it might be removed. However, be cautious as removing too much data can bias the analysis.")
print("- **Capping/Winsorizing:** The outlier value can be capped at a certain percentile (e.g., 95th or 99th percentile) or a value determined by the IQR method (e.g., upper bound = Q3 + 1.5 * IQR).")
print("- **Transformation:** Transformations like logarithmic or square root might reduce the impact of outliers by making the distribution more symmetrical.")
print("- **Keeping as is:** If the outlier is a valid data point and provides important information, it might be kept, especially if the model being used is robust to outliers (e.g., tree-based models).")

# --- Task 2: Price in Product Sales Dataset ---
plt.figure(figsize=(6, 4))
sns.boxplot(x=df_sales['Price'])
plt.title('Box Plot of Price (Product Sales Dataset)')
plt.xlabel('Price')
plt.show()

print("\n--- Handling Outliers in Price (Product Sales Dataset) ---")
print("Potential Outlier Detected (visually far above the upper whisker).")
print("Handling Methods:")
print("- **Consider Business Context:** Investigate if the high price is for a special or luxury item. If it's a valid but unusual product, keeping it might be important for sales analysis of high-value items.")
print("- **Capping:** Cap the price at a reasonable upper limit if it's deemed an error or an unrepresentative extreme.")
print("- **Transformation:** Logarithmic transformation can help in reducing the skewness caused by high-priced items.")
print("- **Segmentation:** Analyze sales with and without the outlier to see its impact on overall statistics and modeling.")

# --- Task 3: Math Score in Student Grades Dataset ---
plt.figure(figsize=(6, 4))
sns.boxplot(x=df_grades['Math Score'])
plt.title('Box Plot of Math Score (Student Grades Dataset)')
plt.xlabel('Math Score')
plt.show()

print("\n--- Handling Outliers in Math Score (Student Grades Dataset) ---")
print("Potential Outlier Detected (visually below the lower whisker).")
print("Handling Methods:")
print("- **Investigate the Cause:** Try to understand why the student has a significantly lower score. It could be due to absence, learning disability, or other factors. Depending on the reason, a decision on handling can be made.")
print("- **Imputation:** If the low score is suspected to be an error, it might be imputed using the mean, median, or other appropriate imputation techniques.")
print("- **Capping:** Cap the lower score at a reasonable minimum if it's considered an extreme but valid case.")
print("- **Keeping as is:** If the low score is a genuine reflection of the student's performance, it should be kept for accurate analysis of student performance distribution.")
```

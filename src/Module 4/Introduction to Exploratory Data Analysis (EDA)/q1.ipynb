Question 4: Handling Outliers
<br>
Description: Identify outliers in numerical columns using box plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris(as_frame=True)
df = iris.frame

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Create box plots for each numerical column
num_cols = len(numerical_cols)
rows = (num_cols + 1) // 2  # Determine the number of rows for subplots
cols = min(num_cols, 2)      # Determine the number of columns for subplots

plt.figure(figsize=(15, 5 * rows))  # Adjust figure height based on the number of rows

for i, col in enumerate(numerical_cols):
    plt.subplot(rows, cols, i + 1)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()

# Identify outliers using IQR method
def identify_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    return outliers

print("\n--- Identifying Outliers using IQR Method ---")
for col in numerical_cols:
    outliers = identify_outliers_iqr(df[col])
    if not outliers.empty:
        print(f"\nOutliers in column '{col}':")
        print(outliers)
    else:
        print(f"\nNo outliers found in column '{col}' using IQR method.")
```

Question 6: Data Transformation
<br>
Description: Transform a categorical column into numerical using Label Encoding.

```python
# Write your code from here

import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder

# Load the Iris dataset
iris = load_iris(as_frame=True)
df = iris.frame

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Select the categorical column to transform
categorical_column = 'target'

# Fit and transform the categorical column
df['target_encoded'] = label_encoder.fit_transform(df[categorical_column])

# Display the original and encoded categorical column
print("--- Original and Label Encoded Categorical Column ---")
print(df[[categorical_column, 'target_encoded']].head())

# Display the mapping of original labels to encoded values
print("\n--- Label Encoding Mapping ---")
mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print(mapping)
```

Question 9: Feature Engineering
<br>
Description: Create a new feature by combining or transforming existing features.

```python
# Write your code from here


import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris(as_frame=True)
df = iris.frame

# Display the first few rows of the original DataFrame
print("--- Original DataFrame ---")
print(df.head())

# 1. Combining two existing numerical features (e.g., ratio of sepal length to sepal width)
df['sepal_ratio'] = df['sepal length (cm)'] / df['sepal width (cm)']
print("\n--- DataFrame with 'sepal_ratio' feature ---")
print(df[['sepal length (cm)', 'sepal width (cm)', 'sepal_ratio']].head())

# 2. Transforming an existing numerical feature (e.g., creating a squared petal length feature)
df['petal_length_squared'] = df['petal length (cm)'] ** 2
print("\n--- DataFrame with 'petal_length_squared' feature ---")
print(df[['petal length (cm)', 'petal_length_squared']].head())

# 3. Combining information from multiple columns to create a categorical feature
# For example, creating a 'petal_area_category' based on petal length and width
df['petal_area'] = df['petal length (cm)'] * df['petal width (cm)']

def categorize_petal_area(area):
    if area < 2.0:
        return 'Small'
    elif 2.0 <= area < 5.0:
        return 'Medium'
    else:
        return 'Large'

df['petal_area_category'] = df['petal_area'].apply(categorize_petal_area)
print("\n--- DataFrame with 'petal_area' and 'petal_area_category' features ---")
print(df[['petal length (cm)', 'petal width (cm)', 'petal_area', 'petal_area_category']].head())
print("\n--- Value counts for 'petal_area_category' ---")
print(df['petal_area_category'].value_counts())
```

Question 10: Advanced Outlier Detection
<br>
Description: Use the Z-score method to identify and handle outliers.

```python
# Write your code from here

import pandas as pd
from scipy import stats
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris(as_frame=True)
df = iris.frame

# Display the original DataFrame info
print("--- Original DataFrame Info ---")
df.info()

# Define a function to identify outliers using the Z-score method
def identify_outliers_zscore(series, threshold=3):
    """Identifies outliers in a pandas Series using the Z-score method."""
    z = np.abs(stats.zscore(series))
    return series[z > threshold]

# Define a function to handle outliers (e.g., by capping them)
def cap_outliers_zscore(series, threshold=3):
    """Caps outliers in a pandas Series to the threshold Z-score values."""
    z = np.abs(stats.zscore(series))
    upper_bound = series.mean() + threshold * series.std()
    lower_bound = series.mean() - threshold * series.std()
    capped_series = series.copy()
    capped_series[series > upper_bound] = upper_bound
    capped_series[series < lower_bound] = lower_bound
    return capped_series

# Select numerical columns (excluding the encoded 'target' column)
numerical_cols = df.select_dtypes(include=['number']).columns.drop('target', errors='ignore')

print("\n--- Outlier Detection using Z-score (Threshold=3) ---")
for col in numerical_cols:
    outliers = identify_outliers_zscore(df[col])
    if not outliers.empty:
        print(f"\nOutliers in column '{col}':")
        print(outliers)
    else:
        print(f"\nNo significant outliers found in column '{col}' using Z-score.")

# Handle outliers by capping them
df_capped = df.copy()
for col in numerical_cols:
    df_capped[col + '_capped'] = cap_outliers_zscore(df[col])

print("\n--- DataFrame with Capped Outliers (Z-score Threshold=3) ---")
print(df_capped.head())
print("\n--- Info of DataFrame with Capped Outliers ---")
df_capped.info()

# You can also remove outliers if appropriate for your analysis
def remove_outliers_zscore(df, columns, threshold=3):
    """Removes rows containing Z-score outliers for specified columns."""
    df_filtered = df.copy()
    initial_rows = len(df_filtered)
    for col in columns:
        z = np.abs(stats.zscore(df_filtered[col]))
        df_filtered = df_filtered[z <= threshold]
    removed_rows = initial_rows - len(df_filtered)
    print(f"\n--- Removed {removed_rows} rows containing Z-score outliers (Threshold={threshold}) ---")
    return df_filtered

df_no_outliers = remove_outliers_zscore(df.copy(), numerical_cols)
print("\n--- Info of DataFrame after Removing Outliers ---")
df_no_outliers.info()


```

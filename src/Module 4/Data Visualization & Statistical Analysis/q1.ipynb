Title: Histogram of a Single Feature <br>

Question 1: Create a histogram for the age feature from a dataset. Interpret what the shape of the histogram tells us about the distribution of the age feature.

```python
# Write your code from here

import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset: Replace this with your actual dataset
data = pd.DataFrame({
    'age': [22, 25, 30, 22, 45, 33, 31, 40, 29, 30, 22, 34, 28, 45, 36, 23, 30, 39, 41, 50]
})

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(data['age'], bins=8, edgecolor='black', color='skyblue')
plt.title('Histogram of Age Feature')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

Title: Descriptive Statistical Analysis<br>

Question 6: Calculate the mean, median, standard deviation, skewness, and kurtosis for the
temperature feature and discuss the results.

```python
# Write your code from here

# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset: height (in cm), weight (in kg), age (in years)
data = {
    'height': [160, 170, 180, 175, 165, 155, 172],
    'weight': [55, 70, 80, 75, 60, 50, 72],
    'age':    [25, 34, 45, 40, 28, 22, 38]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Compute correlation matrix
correlation_matrix = df.corr()

# Set the plot size
plt.figure(figsize=(8, 6))

# Generate the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()

```

Title: Covariance and Correlation between Two Features<br>

Question 7: Compute the covariance and correlation between price and demand in a dataset.
Explain what these metrics indicate.

```python
# Write your code from here

import pandas as pd

# Sample dataset with price and demand
data = {
    'price': [10, 12, 14, 16, 18, 20],
    'demand': [100, 95, 90, 80, 70, 60]  # Demand decreases as price increases
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate covariance
covariance = df['price'].cov(df['demand'])

# Calculate correlation
correlation = df['price'].corr(df['demand'])

# Print results
print(f"Covariance between price and demand: {covariance}")
print(f"Correlation between price and demand: {correlation}")

```

Title: Principal Component Analysis (PCA)<br>

Question 9 : Apply PCA on a dataset with multiple features (e.g., x1 , x2 , x3 , x4 ) and reduce it to two principal components. Visualize the data in the new feature space.

```python
# Write your code from here

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Sample dataset with 4 features
data = {
    'x1': [2.5, 0.5, 2.2, 1.9, 3.1, 2.3, 2.0, 1.0, 1.5, 1.1],
    'x2': [2.4, 0.7, 2.9, 2.2, 3.0, 2.7, 1.6, 1.1, 1.6, 0.9],
    'x3': [1.5, 2.3, 1.8, 2.7, 3.1, 2.2, 1.3, 2.0, 2.1, 1.9],
    'x4': [0.9, 1.2, 0.8, 1.5, 1.9, 1.4, 0.7, 1.0, 1.1, 1.0]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Create a new DataFrame for PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

# Plot the PCA results
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c='skyblue', edgecolors='black')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - 2D Projection of 4D Data')
plt.grid(True)
plt.show()

```

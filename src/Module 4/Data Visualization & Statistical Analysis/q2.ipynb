**Bivariate Analysis (Relationships Between Features)**<br>

Objective: Explore relationships between two features using scatter plots and correlation heatmaps.

Title: Scatter Plots<br>

Task 1: Create a scatter plot between sepal length and sepal width.<br>
Task 2: Scatter plot between petal length and petal width.<br>
Task 3: Scatter plot between sepal length and petal length.<br>

```python
# Write your code from here
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Set the plot style
sns.set(style="whitegrid")

# Task 1: Scatter plot between sepal length and sepal width
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', color='blue')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()

# Task 2: Scatter plot between petal length and petal width
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', color='green')
plt.title('Scatter Plot: Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()

# Task 3: Scatter plot between sepal length and petal length
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', color='purple')
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

Title: Correlation Heatmaps<br>

Task 1: Generate a correlation heatmap of the dataset.<br>
Task 2: Highlight correlation between sepal length and petal length.<br>
Task 3: Highlight correlation between petal width and petal length.<br>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load Iris dataset
iris = sns.load_dataset('iris')

# Select only numeric columns
numeric_cols = iris.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix on numeric columns only
corr = numeric_cols.corr()

# Task 1: Plot correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()

# Task 2 & 3: Highlight specific correlations

# Extract correlations to highlight
corr_sl_pl = corr.loc['sepal_length', 'petal_length']
corr_pw_pl = corr.loc['petal_width', 'petal_length']

print(f"Correlation between Sepal Length and Petal Length: {corr_sl_pl:.2f}")
print(f"Correlation between Petal Width and Petal Length: {corr_pw_pl:.2f}")

# Optional: Visualize heatmap with only these two correlations highlighted

# Create a mask to hide all but the two correlations and their symmetric positions
mask = np.ones_like(corr, dtype=bool)

# Unmask positions of interest
mask[numeric_cols.columns.get_loc('sepal_length'), numeric_cols.columns.get_loc('petal_length')] = False
mask[numeric_cols.columns.get_loc('petal_length'), numeric_cols.columns.get_loc('sepal_length')] = False

mask[numeric_cols.columns.get_loc('petal_width'), numeric_cols.columns.get_loc('petal_length')] = False
mask[numeric_cols.columns.get_loc('petal_length'), numeric_cols.columns.get_loc('petal_width')] = False

plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", mask=mask, cbar=False, square=True)
plt.title('Highlighted Correlations in Iris Dataset')
plt.show()
```

**Statistical Analysis in EDA**<br>

Objective: Calculate basic statistical metrics and explore the relationship between features using correlation and covariance.<br>

Title: Descriptive Statistics<br>

Task 1: Calculate mean, median, and standard deviation of petal length.<br>
Task 2: Calculate skewness and kurtosis of sepal width.<br>
Task 3: Calculate mean, median, and standard deviation of sepal length.

```python
# Write your code from here
import seaborn as sns
import pandas as pd
from scipy.stats import skew, kurtosis

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Task 1: Mean, median, std dev of petal length
petal_length = iris['petal_length']
mean_petal = petal_length.mean()
median_petal = petal_length.median()
std_petal = petal_length.std()

print(f"Petal Length - Mean: {mean_petal:.2f}, Median: {median_petal:.2f}, Std Dev: {std_petal:.2f}")

# Task 2: Skewness and kurtosis of sepal width
sepal_width = iris['sepal_width']
skew_sepal = skew(sepal_width)
kurt_sepal = kurtosis(sepal_width)

print(f"Sepal Width - Skewness: {skew_sepal:.2f}, Kurtosis: {kurt_sepal:.2f}")

# Task 3: Mean, median, std dev of sepal length
sepal_length = iris['sepal_length']
mean_sepal = sepal_length.mean()
median_sepal = sepal_length.median()
std_sepal = sepal_length.std()

print(f"Sepal Length - Mean: {mean_sepal:.2f}, Median: {median_sepal:.2f}, Std Dev: {std_sepal:.2f}")
```

Title: Correlation & Covariance<br>

Task 1: Compute correlation between sepal length and petal length.<br>
Task 2: Compute covariance between petal width and sepal width.<br>
Task 3: Determine the most correlated pair of features.<br>

```python
# Write your code from here
import seaborn as sns
import pandas as pd
import numpy as np

# Load Iris dataset
iris = sns.load_dataset('iris')

# Select numeric features only
numeric_cols = iris.select_dtypes(include=[np.number])

# Task 1: Correlation between sepal length and petal length
corr_sl_pl = numeric_cols['sepal_length'].corr(numeric_cols['petal_length'])
print(f"Correlation between Sepal Length and Petal Length: {corr_sl_pl:.2f}")

# Task 2: Covariance between petal width and sepal width
cov_pw_sw = numeric_cols['petal_width'].cov(numeric_cols['sepal_width'])
print(f"Covariance between Petal Width and Sepal Width: {cov_pw_sw:.4f}")

# Task 3: Find the most correlated pair of features (excluding self-correlation)
corr_matrix = numeric_cols.corr().abs()

# Replace diagonal with 0 to ignore self-correlations
np.fill_diagonal(corr_matrix.values, 0)

# Find max correlation value and its location
max_corr = corr_matrix.max().max()
max_pair = np.where(corr_matrix == max_corr)

feature_1 = corr_matrix.index[max_pair[0][0]]
feature_2 = corr_matrix.columns[max_pair[1][0]]

print(f"Most correlated pair of features: '{feature_1}' and '{feature_2}' with correlation {max_corr:.2f}")
```
